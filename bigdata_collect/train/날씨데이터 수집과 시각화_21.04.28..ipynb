```python
import requests  # 웹 페이지의 html을 가져오는 모듈
from bs4 import BeautifulSoup # html을 파싱하는 모듈

# 웹 페이지를 가져온 뒤 BeautifulSoup 객체로 만듦
response = requests.get("http://www.weather.go.kr/weather/observation/currentweather.jsp")
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
table = soup.find('table', {'class': 'table_develop3'})
```

```python
table.findAll("tr")
```

```python
for tr in table.find_all("tr") :
    tds = tr.find_all("td")
    print("tds = ", tds)
    print("="*100)
```

```python
data = []
```

```python
for tr in table.findAll("tr") :
    tds = tr.findAll("td")
    if len(tds) > 0 :
        print("tds[0](지점) = ", tds[0].text)
        print("tds[5](온도) = ", tds[5].text)
        print("tds[10](습도) = ", tds[10].text)
        data.append([tds[0].text, tds[5].text, tds[10].text])
        print("@"*100)
        print("data = ", data)
        print("@"*100)
    print("=" * 100)
```

```python
for tr in table.findAll("tr") :
    tds = tr.findAll("td")
    if len(tds) > 0 :
        data.append([tds[0].text, tds[5].text, tds[10].text])
```

```python
data
```

```python
with open("weather2.csv", "w") as file :
    file.write('point,temp,hum')
    for item in data :
        row1 = ",".join(item)
        file.write(row1 + "\n")
```
