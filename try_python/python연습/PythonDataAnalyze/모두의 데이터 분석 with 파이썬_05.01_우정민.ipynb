```python
import matplotlib.pyplot as plt # matplotlib.pyplot객체를 plt로 정의하고 호출
import numpy as np # numpy객체를 np로 정의하고 호출
t = np.arange(0., 5., 0.2) # 0 이상 5 미만까지 0.2 간격으로 배열을 만듦
# x = t, y = t 일때 레드 점선 그래프 / x = t, y = t**2 일때 블루 스퀘어점 그래프 / x = t, y = t**3 일때 그린 세모점 그래프
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^') 
plt.show()
```

```python
import matplotlib.pyplot as plt
t = []
p2 = []
p3 = []
for i in range(0, 50, 2) :
    t.append(i / 10)
    p2.append((i / 10) ** 2)
    p3.append((i / 10) ** 3)
print(t)
print(p2)
print(p3)
plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10, 100, 200) # randint(a,b,n) : a 이상 b 미만인 정수 n개를 만드는 함수
y = np.random.randint(10, 100, 200)
# size는 x, y와 같은 크기여야한다.
size = np.random.rand(len(x)) * 100    # np.random.rand(n) : 0 ~ 1 사이에 있는 n개의 실수를 만드는 함수

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000) # randint(a,b,n) : a 이상 b 미만인 정수 n개를 만드는 함수
y = np.random.randint(-100, 100, 1000)
size = np.random.rand(100) * 100    # np.random.rand(n) : 0 ~ 1 사이에 있는 n개의 실수를 만드는 함수

mask1 = abs(x) > 50 # x에 저장된 값 중 절댓값이 50보다 큰 값
mask2 = abs(y) > 50 # y에 저장된 값 중 절댓값이 50보다 큰 값
x = x[mask1 + mask2] # mask1, 2 중 하나라도 만족하는 값 저장
y = y[mask1 + mask2] # mask1, 2 중 하나라도 만족하는 값 저장

print(len(x), len(y))
print(x)
print(mask1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000) # randint(a,b,n) : a 이상 b 미만인 정수 n개를 만드는 함수
y = np.random.randint(-100, 100, 1000)

mask1 = abs(x) > 50 # x에 저장된 값 중 절댓값이 50보다 큰 값
mask2 = abs(y) > 50 # y에 저장된 값 중 절댓값이 50보다 큰 값
x = x[mask1 + mask2] # mask1, 2 중 하나라도 만족하는 값 저장
y = y[mask1 + mask2] # mask1, 2 중 하나라도 만족하는 값 저장

# size 때문에 오류남... x, y와 같은 크기여야...
size = np.random.rand(len(x)) * 100    # np.random.rand(n) : 0 ~ 1 사이에 있는 n개의 실수를 만드는 함수

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
