```python
import csv

f= open('ta_20210419175759.csv')
data = csv.reader(f)
next(data)
result_h = [] #      
result_l = []

for row in data :
    try:
        if row[-1] != '' :  #     
            if 1994 <= int(row[0].split('-')[0]) :
                if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '15' :
                    result_h.append(float(row[-1])) # result     
    except:
        pass
    try:
        if row[-2] != '' :  #     
            if 1994 <= int(row[0].split('-')[0]) :
                if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '15' :
                    result_l.append(float(row[-2])) # result     
    except:
        pass

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.title('    ')
plt.rcParams['axes.unicode_minus'] = False
plt.plot(result_h, 'hotpink', label='')
plt.plot(result_l, 'skyblue', label='')
plt.legend()
plt.show()
```
