### 분수찾기

```python
# 몇번째 그룹에 X가 속하나?
n = int(input())
line = 1
while n > line :
    n-=line
    line+=1

print(line, n)
```

```python
# 조건에 맞게 출력!
if line % 2 == 0:
    print(n, line-n+1, sep='/')
else :
    print(line-n+1, n, sep='/')
```

```python
# 추가 방법
n = int(input())

line = 0 # 라인
max_num = 0 # 입력된 숫자의 라인에서 가장 큰 숫자

while n > max_num :
    line+=1
    max_num +=line
    print(line, max_num)
```

```python
gap = max_num - n
if line%2 == 0 :
    top = line - gap
    under = gap + 1
else :
    under = line - gap
    top = gap + 1
print(f'{top}/{under}')
```
