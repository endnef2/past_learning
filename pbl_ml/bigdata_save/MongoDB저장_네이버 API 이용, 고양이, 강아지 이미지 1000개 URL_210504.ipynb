```python
def get_api_result(keyword, display, start):
    url = "https://openapi.naver.com/v1/search/image?query="\
          + keyword + "&display="+str(display)+\
           "&start="+str(start)
    print("url=", url)
    result = requests.get(url, headers=api_id_pw)
    return result.json()
```

```python
def call_and_print(keyword, total_page=10):
    link_list = []
    for page in range(1, total_page+1):
        display = 100
        start = ((page-1)*display)+1
        print("page = ", page)
        print("start = ", start)
        json_obj = get_api_result(keyword, display, start)
        for item in json_obj["items"]:
            print("item['link'] = ", item['link'])
            link_list.append(item["link"])
        print("="*30)
    return link_list
```

# 고양이 강아지 이미지 1000개씩 MongoDB에 저장

```python
# mongdb에 python_test 데이터 베이스에 접속
db = MongoClient().python_test
```

```python
# python_test에 파일을 저장할 GridFS객체 생성
fs = GridFS(db)
```

```python
# 이미지 파일을 저장할 객체 생성
bucket = GridFSBucket(db)
```

```python
# link1 : 고양이 이미지를 다운로드 할 URL 1000개가 저장된 리스트
for url in link1 :
    try : # 오류에 대한 예외처리
        # url에서 고양이 이미지를 다운 받아서 변수 image에 저장
        image = urllib.request.urlopen(url).read()
        
        # url을 '/'을 기준으로 분리하고 마지막의 이미지파일명과 확장자를 리턴
        image_name = url.split("/")[-1]
        
        # image_name 에서 ?가 있으면(없는 경우 -1이 나옴)
        if image_name.find("?") != -1 :
            # image_name을 ?을 기준으로 분리하고 ?의 첫번째 문자열을 image_type에 대입
            image_name = image_name.split("?")[0]
            
        # 이미지의 기본 타입
        image_type = "jpg"
        
        # image_name에서 .을 포함하고 있으면
        if image_name.find(".") != -1 :
            # image_name(고양이 이미지의 파일명과 확장자)을 . 기준으로 분리하고 마지막 문자열 즉 이미지의 확장자 리턴
            image_type = image_name.split(".")[-1]
            
        # image/이미지 의 확장자를 content_type에 대입
        content_type = "image/{}".format(image_type)
        
        # 이미지 파일의 정보를 저장한 grid_in 생성
        grid_in  = bucket.open_upload_stream(
            image_name, # 이미지 파일의 이름
            metadata = {"contentType":content_type, "type":"cat"} # 이미지 파일 타입
         )
        # 이미지의 내용을 GridFS에 저장
        grid_in.write(image)
        # 이미지 저장 종료
        grid_in.close()
        print("image_type :", image_type, ": content_type : ", content_type, ": image_name :", image_name)
    except :
        print("에러 발생")
```
