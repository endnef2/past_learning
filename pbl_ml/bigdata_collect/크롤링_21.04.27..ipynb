```python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20210427"
```

```python
html = urllib.request.urlopen(url)
```

```python
soup= BeautifulSoup(html, "html.parser")
```

```python
soup
```

```python
titleOfMovie1 = soup.findAll("td", {"class":"title"})
```

```python
titleOfMovie1
```

```python
for item in titleOfMovie1 :
    print("item of text = ", item.text)
    print("@"*100)
    print("item of text.strip() = ", item.text.strip())
    print("="*100)
    print("="*100)
```

```python
for item in titleOfMovie1 :
    print("item : ", item.text.strip())
    print("-"*100)
```

```python
points = soup.findAll("td", {"class":"point"})
```

```python
points
```

```python
len(titleOfMovie1)
```

```python
len(points)
```

```python
for i in range(len(points)):
    print("i = ", i)
```

```python
for i in range(len(points)):
    print("i = ", i)
    print("titleOfMovie1[i] = ", titleOfMovie1[i].text.strip())
    print("points[i] = ", points[i].text.strip())
    print("="*100)
```

```python
for title, star in zip(titleOfMovie1, points) :
    print("title.text.strip() = ", title.text.strip(), "star.text.strip() = ", star.text.strip())
    print("="*100)
```

## zip 함수

```python
list1 = ["a", "b", "c"]
list1
```

```python
list2 = [100, 200, 300]
list2
```

```python
zip(list1, list2)
```

```python
for item in zip(list1, list2) :
    print("item : ", item)
```

```python
for data1, data2 in zip(list1, list2) :
    print("data1 : ", data1)
    print("data2 : ", data2)
```
