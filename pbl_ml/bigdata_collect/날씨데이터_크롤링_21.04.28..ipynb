```python
from bs4 import BeautifulSoup
```

```python
response = requests.get("http://www.weather.go.kr/weather/observation/currentweather.jsp")
soup = BeautifulSoup(response.text, 'html.parser') #BeautifulSoup 써야오류안나...
```

```python
table = soup.find("table", {"class": "table_develop3"})
```

```python
table.findAll("tr")
```

```python
for tr in table.find_all("tr") :
    print("tr = ", tr)
    print("=" * 100)
```

```python
for tr in table.findAll("tr") :
    tds = tr.findAll("td")
    print("tds = ", tds)
    print("=" * 100)
# 0 : 동네, 5 : 현재기온, 10 : 습도
```

```python
for tr in table.findAll("tr") :
    tds = tr.findAll("td")
    print("tds = ", tds)
    print("*" * 100)
    print("tds[0] = ", tds[0])
    print("tds[5] = ", tds[5])
    print("tds[10] = ", tds[10])
```

```python
for tr in table.findAll("tr") :
    tds = tr.findAll("td")
    print("tds = ", tds)
    print("*" * 100)
    print("len(tds) = ", len(tds))
    print("=" * 100)
```

```python
for tr in table.findAll("tr") :
    tds = tr.findAll("td")
    if len(tds) > 0 :
        print("tds[0](지점) = ", tds[0].text)
        print("tds[5](온도) = ", tds[5].text)
        print("tds[10](습도) = ", tds[10].text)
    print("=" * 100)
```

```python
# join 함수 :  리스트에서 데이터 하나씩 꺼내서 문자열 하나 만들어서 리턴
for item in data:
    print("item : ", item)
    row = " V^^V ".join(item)
    print("row : ", row)
    print("="*100)
```

```python
with open('weather.csv', 'w') as file :
    print("파일 저장 할꺼임")
    file.write('point,temp,hum')
    for item in data :
        print("item = ", item)
        row1 = ",".join(item)
        print("row1 = ", row1)
        file.write(row1 + "\n")
```
