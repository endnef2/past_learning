```python
# 우리 동네에 대한 데이터인지 확인한다
# 우리 동네일 경우 0세부터 100세 이상까지의 인구수를 순서대로 저장한다.
import csv
f = open('age.csv')
data = csv.reader(f)
header = next(data)
result = []
for row in data:
    if '금정동' in row[0]:
        for i in row[3:] :
            result.append(int(i))
print(result)
print(len(result))
```

```python
# 아름동의 인구구조 시각화 결과를 얻어라!
import csv
f = open('age.csv')
data = csv.reader(f)
header = next(data)
result = []
aaa = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
for row in data:
    if aaa in row[0]:
        for i in row[3:] :
            result.append(int(i.replace(',','')))  # age.csv파일의 콤마(,)를 빈문자로 바꿔줌
            
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(aaa + ' 지역의 인구 구조')
plt.plot(result)
plt.show()
```

```python
# 항아리 모양 그래프 그리기

import csv
f = open('gender.csv') # csv파일 불러오기
data = csv.reader(f)

# 데이터를 넣을 리스트 만들기
m = []
f = []

name = input('지역의 이름을 알려주세요 : ')

for row in data:
    if name in row[0]: # in row 쓰기...
        for i in row[3:104] :
            m.append(-int(i.replace(',',''))) # m에 해당하는 데이터를 음수로 넣기( , -> 공백으로!)
        for i in row[106:] :
            f.append(int(i.replace(',',''))) # f에 해당하는 데이터 넣기

# 그래프 만들기

import matplotlib.pyplot as plt

# plt.style.use('ggplot')
# plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend() # 범례 표현
plt.show()
```

```python
import csv
f = open('gender.csv') # csv파일 불러오기
data = csv.reader(f)
size = []

name = input('지역의 이름을 알려주세요 : ')

for row in data:
    if name in row[0]: # in row 쓰기...
        m = 0
        f = 0
        for i in range(101) :
            m += int(row[i+3].replace(',','')) # 3 ~ 103번까지 데이터 더하기(, =>공백으로 바꾸기)
            f += int(row[i+106].replace(',','')) # 106 ~ 206번까지 데이터 더하기
        
        break # 반복문 실행 멈추기

size.append(m)
size.append(f)
print(size)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family='Malgun Gothic') # 한글 폰트
color = ['crimson', 'darkcyan'] #색 효과
plt.axis('equal') # 동그란 원 그리기
plt.pie(size, labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90) # 레이블 추가, 소수점 아래 첫번째 소수점까지 비율표현 및 범례추가
plt.title(name+' 지역의 남녀 성별 비율') # 제목 추가
plt.show()
```
