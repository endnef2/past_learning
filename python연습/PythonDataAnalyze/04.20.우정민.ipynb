```python
import csv

f= open('ta_20210419175759.csv')
data = csv.reader(f)
next(data)
result = [] # 최고 기온 데이터를 저장할 리스트 생성

for row in data :
    try:
        if row[-1] != '' :  # 최고 기온 데이터 값이 존재한다면
            if 1994 <= int(row[0].split('-')[0]) :
                if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '15' :
                    result.append(float(row[-1])) # result 리스트에 최고 기온 값 추가
    except:
        pass
import matplotlib.pyplot as plt
plt.plot(result, 'hotpink')
plt.show()
```

```python
# 내 생일 이후 매년 최고, 최저기온 데이터 시각화
```

```python
import csv

f= open('ta_20210419175759.csv')
data = csv.reader(f)
next(data)
result_h = [] # 최고 기온 데이터를 저장할 리스트 생성
result_l = []

for row in data :
    try:
        if row[-1] != '' :  # 최고 기온 데이터 값이 존재한다면
            if 1994 <= int(row[0].split('-')[0]) :
                if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '15' :
                    result_h.append(float(row[-1])) # result 리스트에 최고 기온 값 추가
    except:
        pass
    try:
        if row[-2] != '' :  # 최고 기온 데이터 값이 존재한다면
            if 1994 <= int(row[0].split('-')[0]) :
                if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '15' :
                    result_l.append(float(row[-2])) # result 리스트에 최고 기온 값 추가
    except:
        pass

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.title('내 생일의 기온 변화 그래프')
plt.rcParams['axes.unicode_minus'] = False
plt.plot(result_h, 'hotpink', label='최고기온')
plt.plot(result_l, 'skyblue', label='최저기온')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f= open('ta_20210419175759.csv')
data = csv.reader(f)
next(data)
aug = [] # 최고 기온 데이터를 저장할 리스트 생성

for row in data :
    try:
        month = row[0].split('-')[1]
        if row[-1] != '' :  # 최고 기온 데이터 값이 존재한다면
            if month == '08' :
                aug.append(float(row[-1])) # result 리스트에 최고 기온 값 추가
    except:
        pass
    
plt.hist(aug, bins=100, color='r')
plt.show()
```
